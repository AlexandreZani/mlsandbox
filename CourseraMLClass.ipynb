```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

# First we need to generate some data
np.random.seed(0)
DATA_SZ = 500
X_train = np.random.rand(2*DATA_SZ)-0.5
X_train = np.append(np.ones(shape=(DATA_SZ,1)), X_train)
X_train = np.reshape(X_train, newshape=(3, DATA_SZ)).T

model = np.reshape(np.array([ 0.4461758 , -0.26692635,  0.41031005]), newshape=(3, 1))
y_train = X_train.dot(model)
y_train = np.reshape((np.random.rand(DATA_SZ)-0.5)*0.0001, newshape=(DATA_SZ,1)) + y_train

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_train[:, 1], X_train[:, 2], y_train)

plt.show()
```
